## "Exploratory Data Analysis"

```python
ax=sns.countplot(x= 'Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

###### 7832 females as compared to 3407 males make purchase

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Expenditure'].sum().sort_values(by='Expenditure', ascending=False)
sns.barplot(x='Gender', y='Expenditure', data=sales_gen)
```

```python
ax=sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

###### Comparison of age groups making purchases. Maximum purchase is done by females of age group 26-35

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

###### Maximum number of purchases are made by people of Uttar Pradesh, Maharashtra and Karnataka

```python
sales_state=df.groupby(['State'], as_index=False)['Expenditure'].sum().sort_values(by='Expenditure', ascending=False).head(10)
sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(data=sales_state, x='State', y='Expenditure')
```

###### Maximum amount of purchases are made by people of Uttar Pradesh, Maharashtra and Karnataka

```python
ax=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize': (7, 4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

###### Maximum number of purcahses are made by married people

```python
sales_state=df.groupby(['Marital_Status','Gender'], as_index=False)['Expenditure'].sum().sort_values(by='Expenditure', ascending=False).head(10)
sns.set(rc={'figure.figsize': (6, 5)})
sns.barplot(data=sales_state, x='Marital_Status', y='Expenditure', hue='Gender')
```

```python
sns.set(rc={'figure.figsize': (20, 5)})
ax=sns.countplot(data=df, x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

###### Maximum purchase made by people in IT, healthcare and aviation sector

```python
sales_state=df.groupby(['Occupation'], as_index=False)['Expenditure'].sum().sort_values(by='Expenditure', ascending=False).head(10)
sns.barplot(data=sales_state, x='Occupation', y='Expenditure')
```

```python
sns.set(rc={'figure.figsize': (20, 5)})
ax=sns.countplot(data=df, x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

###### From above graph we can conclude that most purchased products are Clothing and Apparel, Food and Electronics and Gadgets

```python
sales_state=df.groupby(['Product_Category'], as_index=False)['Expenditure'].sum().sort_values(by='Expenditure', ascending=False).head(10)
sns.barplot(data=sales_state, x='Product_Category', y='Expenditure')
```

```python
sales_state=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=sales_state, x='Product_ID', y='Orders')
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

# CONCLUSION

##### Married women age group 26-35 years from UP, Maharashtra and Karnataka working in IT, Healthcare and aviation are more likely to buy products from Clothing, Food and Electronics category.
